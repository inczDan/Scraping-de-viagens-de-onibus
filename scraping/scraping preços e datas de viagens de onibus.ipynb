## Destrinchando as tags


```python
#se for unidade, conseguimos pegar o dado mais profundo (limpo)
data = soup.find('p', class_=re.compile('ird-dia')).find('span')
data.text
```

```python
#se for toda pagina, pegamos apenas a base
datas = soup.find_all('p', class_=re.compile('ird-dia'))
datas
```

```python
lista_datas = []
for dia in datas:
    dado_dia = dia.find('span').text.split(',')
    dado_dia_limpo = dado_dia[0].replace('[', '').replace(']','').replace('/','-')
    lista_datas.append(dado_dia_limpo)
for item in lista_datas:
    print(item)
    print('--------')
```

## pega preço

```python
table_preco = soup.find_all('div', class_=re.compile('p-preco'))
table_preco
```

```python
lista_preco = []
for preco in table_preco:
    dado_preco = preco.text.replace('R$', '').replace(',', '.').strip()
    lista_preco.append(dado_preco)
print(lista_preco)
```
